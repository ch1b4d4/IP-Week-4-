```python
#Changing the Data Types of some values in objects to appropriate data types

#1. City
# df['city'].astype(str)
df['city'] = df['city'].astype('string', '|S')
df['charging_status'] = df['charging_status'].astype('string', '|S')
df['subscription_status'] = df['subscription_status'].astype('string', '|S')
df['address'] = df['address'].astype('string', '|S')
df['charging_status'] = df['charging_status'].astype('string', '|S')
df['charging_status'] = df['charging_status'].astype('string', '|S')
df['city'] = df['city'].astype('string', '|S')
df['id'] = df['id'].astype('string', '|S')
df['kind'] = df['kind'].astype('string', '|S')
df['geo_point'] = df['geo_point'].astype('string', '|S')
df['public_name'] = df['public_name'].astype('string', '|S')
df['rental_status'] = df['rental_status'].astype('string', '|S')
df['station_type'] = df['station_type'].astype('string', '|S')
df['status'] = df['status'].astype('string', '|S')
```

```python
df.info()
```

```python
#Last attempt to check for null values in the dataset
df.isnull().any()
```

```python
#Drop the column cars as it potrays same data as the blue car counter
clean = df.drop('cars', axis = 1)
clean.head(2)
```

```python
#Removing of white spaces from the dataset
clean.columns = clean.columns.str.strip().str.lower().str.replace(' ', '_')
clean.head(2)
```

```python
#Dropping all rows where the value of "Bluecar counter" is less than 0 or greater than 7
blue_clean = clean[(clean['bluecar_counter'] < 0) & (clean['bluecar_counter'] > 7) ].index
clean.drop(blue_clean , inplace=True)
# clean.head(2)
```

```python
#Dropping all rows where the value of "utlib counter" is less than 0 or greater than 4
utlib_clean = clean[(clean['utilib_counter'] < 0) & (clean['utilib_counter'] > 4) ].index
clean.drop(utlib_clean , inplace=True)
# utlib_clean.shape
# clean.shape
```

```python
#Dropping all rows where the value of "utilib_1.4_counter" is less than 0 or greater than 5
utlib_14_clean = clean[(clean['utilib_1.4_counter'] < 0) & (clean['utilib_1.4_counter'] > 5) ].index
clean.drop(utlib_14_clean , inplace=True)
```

```python
#Dropping all rows where the value of "charge_slots" is less than 0 or greater than 3
charge_clean = clean[(clean['charge_slots'] < 0) & (clean['charge_slots'] > 3) ].index
clean.drop(charge_clean , inplace=True)
```

```python
#Combining the date, month and year columns to make one column with this information
clean['date']=pd.to_datetime(clean[['year','month','day', 'hour', 'minute']])
clean.head()
```

```python
clean = clean.drop(["year","month","day","hour","minute"], axis = 1)
clean.head(2)
```

```python
#Question/Challenge 1

#Identify the most popular hour of the day for picking up a shared electric car 
# (Bluecar) in the city of Paris over the month of April 2018.

#Define a dataset with Paris as the city and the blue car being used at a specific time

paris = clean[clean['city'] == 'Paris']

#Using the Dataset with the city as Paris to see the time cars were always available
#at specific stations to show how popular that time and station was

blue_paris = (paris[["bluecar_counter", "date", "address", "id"]])
paris_b = blue_paris.sort_values(["bluecar_counter", "address", "id", "date"],ascending = True)
paris_b.head()

```
